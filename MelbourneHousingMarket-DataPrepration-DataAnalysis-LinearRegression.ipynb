```python
# number of null values in each columns 
for i in columns:
    count =0
    k = df[i].isnull()
    for m in k:
        if m == True :
            count = count +1
    print(i ," = ",count)
```

```python
df_NoDuplicate["Bathroom"] = df_NoDuplicate["Bathroom"].fillna(1.0)    # using mode
df_NoDuplicate["Bedroom2"] = df_NoDuplicate["Bedroom2"].fillna(3.0)    # using mode
df_NoDuplicate["Car"] = df_NoDuplicate["Car"].fillna(2.0)              # using mode
df_NoDuplicate["Landsize"] = df_NoDuplicate["Landsize"].fillna(2.0)    # using median
df_NoDuplicate["YearBuilt"] = df_NoDuplicate["YearBuilt"].fillna(1970) # using mode
df_NoDuplicate[["Bedroom2","Bathroom","Car","Landsize","YearBuilt"]]
```

```python
# number of null values in each columns 
for i in columns:
    count =0
    k = df_NoDuplicate[i].isnull()
    for m in k:
        if m == True :
            count = count +1
    print(i ," = ",count)
```

```python
df_drop_NanDistance = df_NoDuplicate[~ df_NoDuplicate["Distance"].isnull()]
df_drop_NanDistance
```

```python
df_drop_NanPrice = df_drop_NanDistance[~ df_drop_NanDistance['Price'].isnull()]
df_drop_NanPrice
```

```python
df_NanPrice = df_drop_NanDistance[ df_drop_NanDistance['Price'].isnull()]
df_NanPrice
```

```python
X_train = df_drop_NanPrice[["Distance","Bedroom2" , "Bathroom","Car" , "Landsize" , "YearBuilt"]]
Y_train = df_drop_NanPrice["Price"]
X_test = df_NanPrice[["Distance","Bedroom2" , "Bathroom","Car" , "Landsize" , "YearBuilt"]]


X_train.shape , Y_train.shape , X_test.shape
```

###### creating linear model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
df_NanPrice["Price"] = model.predict(X_test)
```
