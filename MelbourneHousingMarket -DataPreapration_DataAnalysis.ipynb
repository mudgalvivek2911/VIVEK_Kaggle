```python
list_object_col = df_NoDuplicate.select_dtypes("object").columns
```

```python
df_NoDuplicate["Bathroom"] = df_NoDuplicate["Bathroom"].fillna(1.0)    # using mode
df_NoDuplicate["Bedroom2"] = df_NoDuplicate["Bedroom2"].fillna(3.0)    # using mode
df_NoDuplicate["Car"] = df_NoDuplicate["Car"].fillna(2.0)              # using mode
df_NoDuplicate["Landsize"] = df_NoDuplicate["Landsize"].fillna(2.0)    # using median
df_NoDuplicate["YearBuilt"] = df_NoDuplicate["YearBuilt"].fillna(1970) # using mode
df_NoDuplicate[["Bedroom2","Bathroom","Car","Landsize","YearBuilt"]]
```
